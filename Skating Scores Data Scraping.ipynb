```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import csv

url = "https://skatingscores.com/men/"
url1 = "https://skatingscores.com/2122/oly/sr/men/long/"
```

```python
def get_table(path):    
    page = requests.get(path)
    pagetext = page.text
    soup = BeautifulSoup(pagetext,"lxml")
    table = soup.find('table')
    
    list_of_rows = []
    for row in table.findAll('tr'):
        list_of_cells = []
        for cell in row.findAll(["th","td"]):
            text = cell.text
            list_of_cells.append(text)
        list_of_rows.append(list_of_cells)
    
    res = pd.DataFrame(list_of_rows)
    return res

def get_score_breakdown(path):
    page = requests.get(path)
    pagetext = page.text
    soup = BeautifulSoup(pagetext, "lxml")
    
    inhalt_div = soup.find(id="inhalt")
    table_div = inhalt_div.find('div', class_="col-12 col-t-12 col-d-6")
    table = table_div.find_all('table') 
    #type(table)
    list_of_rows = []
    for i in range(len(table)):
        for row in table[i].findAll('tr'):
            list_of_cells = []
            for cell in row.findAll(["th","td"]):
                text = cell.text
                list_of_cells.append(text)
            list_of_rows.append(list_of_cells)
    res = pd.DataFrame(list_of_rows)
    return res
```

```python
get_table(url)
```

```python
get_score_breakdown(url1)
```
